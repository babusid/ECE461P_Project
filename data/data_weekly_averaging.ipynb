```python
import pandas as pd
import numpy as np
```

```python
# Pseudocode

def weekly_averaging(csv_input, csv_output):
    # Read the csv file
    print("Reading csv....")
    data = pd.read_csv(csv_input)
    print("Reading complete!")
    
    # Get the list of fips to loop through 
    fips = np.array(data['fips'])
    fips_unique = np.unique(fips)

    # create an Empty DataFrame object
    df = pd.DataFrame()


    # For each fip in fips_unique:
    for fip in fips_unique:
        print("Working on fip: " + str(fip) + "...")
        # Extract the subset corresponding to the fip/county
        subset = data[fips == fip]

        # Convert the date column to a datetime object
        subset.loc[:,'date'] = pd.to_datetime(subset.loc[:,'date']) 

        # Use grouby & pd.Grouper to get the mean for each week
        weekly_avg = subset.groupby(pd.Grouper(freq='W-TUE', closed='right', label='right', key = 'date')).mean()
        subset = weekly_avg.dropna()


        # Concatenate the subset (dataframe) to a new dataframe
        df = pd.concat([df, subset])

    # Export the pandas dataframe as csv file
    print("Exporting csv file ...")
    df.to_csv(csv_output)
```

```python
#weekly_averaging('train_timeseries.csv', 'train_with_weekly_avg.csv')
#weekly_averaging('test_timeseries.csv', 'test_with_weekly_avg.csv')
weekly_averaging('validation_timeseries.csv', 'validation_with_weekly_avg.csv')
```

One caveat to the code is that the date column is now the first column (so compared to the origin code, the fip and the date column is flipped)
